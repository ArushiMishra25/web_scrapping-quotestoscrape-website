# Finding all info of one page first

```python
#finding all the quotes of one page
quotes=[]
for i in soup.find_all('span',class_='text'):
    quotes.append(i.text)
```

```python
quotes
```

```python
#finding all the authors
authors=[]
for i in soup.find_all('small',class_='author'):
    authors.append(i.text)
```

```python
authors
```

```python
#about link of all authors
links=[]
sp=soup.find_all('div',class_='quote')
for i in sp:
    w='https://quotes.toscrape.com'+i.find('a').get('href')
    links.append(w)
```

```python
links
```

```python
#tags 
tags=[]
for i in soup.find_all('div',class_='tags'):
    q=[]
    for j in i.find_all('a',class_='tag'):
        q.append(j.text)
    tags.append(q)
    
```

```python
tags
```

```python
x={}
x['authors']=authors
```

```python
x
```

```python
x['links']=links
```

```python
x['quotes']=quotes
```

```python
x['Tags']=tags
```

```python
df=pd.DataFrame(x)
df.rename(columns={'authors':'Authors','links':'Links','quotes':'Quotes'},inplace=True)
```

# Scrapping from multiple pages

```python
#moving to scrap the other pages like page 2 3 4 ...
```

```python
q=[]
a=[]
l=[]
t=[]
for i in tqdm(range(2,11)):
    link='https://quotes.toscrape.com/page/'+str(i)
    s=requests.get(link)
    soup1=bs(s.text,'html.parser')
    for i in soup1.find_all('span',class_='text'):
        q.append(i.text)
    for i in soup1.find_all('small',class_='author'):
        a.append(i.text)
    sp=soup1.find_all('div',class_='quote')
    for i in sp:
        w='https://quotes.toscrape.com'+i.find('a').get('href')
        l.append(w)
    for i in soup1.find_all('div',class_='tags'):
        q1=[]
        for j in i.find_all('a',class_='tag'):
            q1.append(j.text)
        t.append(q1)
```

```python
q
```

```python
final={}
final['Quote']=q
final['Author']=a
final['Links']=l
final['Tags']=t
final
```

```python
df_f=pd.DataFrame(final)
```
